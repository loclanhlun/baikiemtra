```
# Câu 1:
def RemoveValue(listA):
  listB = []
  for i in range(0, len(listA)):
      if(listA[i] != 10):
        listB.append(listA[i])
  return listB

listA = [1,10,20,5,60,30,10,9,10]
print(RemoveValue(listA))
```

```
#Câu 3: 
def removeKey(person, keyToRemove):
  for key in keyToRemove:
      person.pop(key,None)
  return person
person = {
    "name" : "Quoc Nam",
    "age" : 28,
    "salary" : 8000,
    "city" : "Tuy Hoa"
}
keyToRemove = ["salary" , "city"]
print(removeKey(person, keyToRemove))
```
